```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Setting the LaTeX rendering configuration
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

# Generating data
T = np.arange(20)
function = np.sin(T) + np.random.rand(len(T))

# Setting the index for vertical line
n = 10  # Change this value as needed
vertical_line_x = n

# Plotting
plt.figure(figsize=(20, 6))
plt.scatter(T, function, label=r"$u_n$", s=60)

# Adding a vertical line
plt.axvline(x=vertical_line_x, color='gray', linestyle='--', label=f'$n={n}$')

# Highlighting the maximum and minimum after the vertical line
post_n = T[T >= vertical_line_x]
post_function = function[T >= vertical_line_x]

if len(post_function) > 0:
    max_value = np.max(post_function)
    min_value = np.min(post_function)
    max_index = np.argmax(post_function) + len(T[T < vertical_line_x])
    min_index = np.argmin(post_function) + len(T[T < vertical_line_x])

    plt.scatter(max_index, max_value, color='red', label=r'$\sup_{p \geq n} u_p$', s=100)
    plt.scatter(min_index, min_value, color='green', label=r'$\inf_{p \geq n} u_p$', s=100)

# Customizing the legend
plt.legend()

# Customizing the x-axis
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))


plt.xlabel('X-axis', fontsize=30)  # X-axis label font size
plt.ylabel('Y-axis', fontsize=30)  # Y-axis label font size

# Set tick parameters for bigger x and y tick labels
plt.xticks(fontsize=25)  # X-tick label font size
plt.yticks(fontsize=25)  # Y-tick label font size

# Show legend with increased font size
plt.legend(fontsize=30)

# Show plot
plt.title("")
plt.xlabel(r"n")
plt.ylabel(r"$u_n$")
plt.grid()
plt.tight_layout()
plt.savefig("example_sup_inf.png", dpi=500, bbox_inches='tight')
plt.show()
```
